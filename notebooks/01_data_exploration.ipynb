Sanity-check the FastF1 ingestion: row counts per session, missingness, lap-time distributions.

```python
%load_ext autoreload
%autoreload 2

import pandas as pd
from pitwall.utils.io import read_parquet_glob

laps = read_parquet_glob('laps', year=2024)
print(f'Total lap rows: {len(laps):,}')
laps.head()
```

```python
laps.groupby('CircuitName').size().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
laps['LapTimeSeconds'].dropna().plot(kind='hist', bins=100, figsize=(10, 4))
plt.title('Lap time distribution, 2024 season')
plt.xlabel('Lap time (s)')
plt.show()
```
